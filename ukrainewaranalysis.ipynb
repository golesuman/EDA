```python
df = pd.read_csv('../input/2022-ukraine-russian-war/russia_losses_equipment.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.fillna(0, inplace = True)
```

```python
df.info()
```

```python
df.describe()
```

```python
#losses per day day
sns.set_theme(style = 'whitegrid')
plt.figure(figsize=(20,15))
for i, aircraft in enumerate(df.columns):
    if aircraft != 'date' and aircraft != 'day':
        ax = plt.subplot(5,3, i-1)
        plt.xlabel(aircraft)
        plt.plot(df['day'], df[aircraft])
        plt.subplots_adjust(hspace=0.9)
```

# As we can see that fuel tanks, mobile srbm systems, drones, special equipments were not that used for initial 10 days. And it started growing after that.

```python
plt.figure(figsize = (20,10))sns.heatmap(df.corr(), annot = True, cmap = 'BuPu')
```

```python
# Analysis on the personnel loses
df_personnel = pd.read_csv('../input/2022-ukraine-russian-war/russia_losses_personnel.csv')
df_personnel.sample(10)
```

```python
# dropping the date since day is given
df_personnel.drop('date', inplace = True, axis = 1 )
```

```python
df_personnel.info()
```

```python
df_personnel.describe()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df_personnel.corr(), annot = True, cmap = 'Greens')
```

```python
plt.figure(figsize = (15,10))
for n,feature in enumerate(df_personnel.columns):
    if feature != 'personnel*' and feature != 'day':
        plt.subplot(2,2, n+1)
        plt.plot(df_personnel['day'], df_personnel[feature])
        plt.subplots_adjust(hspace=0.5)
        plt.xlabel(feature)
    
```

### As we can see that the no of personnel losing their lives are increasing as the day increases and the no of war prisoners are also increasing
